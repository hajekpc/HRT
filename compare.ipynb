```python
calib = []

# calib.append(readCalibration("calibration/Tenzo2-08-05-2023-19-24-13.csv"))
calib.append(readCalibration("calibration/Tenzo2-13-05-2023-19-20-01.csv"))
calib.append(readCalibration("calibration/Tenzo2-14-05-2023-12-12-01.csv"))
calib.append(readCalibration("calibration/Tenzo2-14-05-2023-12-28-45.csv"))

ramps = []
for c in calib:
    ramps.append(c.loc[c["Status"].isin(["up","down"])])
    
for r in ramps:
    plt.plot(r["F"], r["V"], ".")


```
